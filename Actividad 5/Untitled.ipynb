```python
import pandas as pd
import numpy as np
import matplotlib as plt
import scipy.stats as stats
from pylab import *
df=pd.read_csv("Mazatlan.csv",header=int(0))
```

```python
from datetime import datetime
df['fecha']= df.apply(lambda x:datetime.strptime("{0} {1} {2} {3}".format(x[u'anio'],x[u'mes'], x[u'dia'], x[u'hora(utc)']), "%Y %m %d %H"),axis=1)
```

```python
x=df['fecha']
y=df['altura(mm)']
```

```python
plt.plot(x,y)
plt.grid(True)

plt.title('Mareas de Mazatlán, Sinaloa (Abril 2017)')
plt.ylabel('Nivel del Mar (mm)')
plt.xlabel('Fecha')
```
