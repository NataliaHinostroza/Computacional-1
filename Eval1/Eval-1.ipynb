```python
import pandas as pd  
import matplotlib as plt
import matplotlib.pyplot as mplt
import numpy as np
df = pd.read_csv("/home/nataliahm/Eval1/Datos.csv", skiprows=2, names=['Year','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','JD','DN','DJF','MAM','JJA','SON'])
df.head()
```

```python
df.Year= pd.to_numeric(df.Year, errors='coerce')
df.JD= pd.to_numeric(df.JD, errors='coerce')
```

```python
df.columns
```

```python
x=df[u'Year']
y=df[u'JD']
```

```python
mplt.plot(x,y, color='k', marker='.')
mplt.grid(True)
mplt.title('Temperatura promedio anual')
mplt.xlabel('Año')
mplt.ylabel('Temperatura (°C)')
```

```python
mplt.show()
```

```python
import numpy as np
import statsmodels.api as sm
from scipy.interpolate import interp1d
```

```python
lowess = sm.nonparametric.lowess(y, x, frac=.3)
x = df.Year
y = df.JD

lowess_x = list(zip(*lowess))[0]
lowess_y = list(zip(*lowess))[1]


f = interp1d(lowess_x, lowess_y, bounds_error=False)

xnew = [i/10. for i in range(140)]


ynew = f(xnew)

mplt.title('Temperatura promedio anual')
mplt.xlabel('Año')
mplt.ylabel('Temperatura (°C)')

mplt.plot(x, y,'.',color='k', label='Media anual', linestyle='-')
mplt.plot(lowess_x, lowess_y, color='R', label='Lowess')
mplt.plot(xnew, ynew)
mplt.legend(loc= 'upper left')
```

```python
mplt.show()
```
